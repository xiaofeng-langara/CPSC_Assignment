## import python packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## load titanic dataset

```python
df=pd.read_csv("../Data/titanic.csv")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#df.fillna(value=df[['Age']].median(),inplace=True)
```

## 1. Determine if the survival rate is associated to the class of passenger

```python
sns.countplot(x='Survived',hue='Pclass',data=df)
```

## 2. Determine if the survival rate is associated to the gender

```python
sns.countplot(x='Survived',hue='Sex',data=df)
```

## 3. Determine the survival rate is associated to the age

```python
sns.histplot(data=df,x='Age',hue='Survived',element='poly')
```
